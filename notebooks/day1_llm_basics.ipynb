## 2. Evolution of AI: From Machine Learning to Large Language Models

Artificial Intelligence has evolved significantly over the past few decades.
Understanding this evolution helps us see how **Large Language Models (LLMs)** emerged.

---

### Machine Learning

Traditional machine learning models learn patterns from structured datasets and use those patterns to make predictions.

Examples of machine learning algorithms include:

- Linear Regression
- Decision Trees
- Random Forest
- Support Vector Machines

These models typically require **manual feature engineering**, where humans must decide which input features are important.

Example applications include:

- Predicting house prices
- Fraud detection
- Customer churn prediction
- Spam classification

Although machine learning models work well for structured data, they struggle with **complex unstructured data such as text, images, and audio**.

---

### Deep Learning

Deep learning introduced **neural networks with many layers**, allowing models to automatically learn complex patterns from data.

Deep learning models significantly improved performance in areas such as:

- Image recognition
- Speech recognition
- Natural language processing

Common deep learning architectures include:

- Convolutional Neural Networks (CNNs) – widely used for images
- Recurrent Neural Networks (RNNs) – used for sequential data
- Long Short-Term Memory networks (LSTMs) – improved sequence modelling

However, early deep learning models for language processing still faced limitations when handling **long text sequences**.

---

### Encoder–Decoder Architecture

Before Transformers, many Natural Language Processing tasks used an **Encoder–Decoder architecture**.

This architecture became popular for tasks such as:

- Machine translation
- Text summarization
- Question answering

The idea is simple:

1. The **Encoder** reads the input sequence and converts it into a numerical representation.
2. The **Decoder** uses this representation to generate the output sequence.

Example:

Input sentence (English):

> "How are you?"

Output sentence (French):

> "Comment ça va?"

However, early encoder–decoder models had an important limitation.
They compressed the entire input sequence into a single representation, which made it difficult to handle **long sentences**.

---

### Attention Mechanism

To overcome this limitation, researchers introduced the **Attention Mechanism**.

Instead of relying on a single compressed representation, attention allows the model to **focus on different parts of the input sequence while generating each output word**.

This means the model can dynamically determine **which words in the input are most relevant at each step**.

The introduction of attention significantly improved performance in many natural language processing tasks.

---

### Transformer Architecture

A major breakthrough occurred in 2017 when researchers introduced the **Transformer architecture** in the paper:

> *"Attention Is All You Need"*

Transformers removed recurrent networks entirely and relied completely on **attention mechanisms**.

Key advantages of Transformers include:

- Parallel processing of sequences
- Better handling of long-range dependencies
- Faster training
- Ability to scale to very large models

The Transformer architecture became the foundation for modern language models.

---

### Large Language Models (LLMs)

Large Language Models are deep neural networks built using the **Transformer architecture** and trained on extremely large text datasets.

These models learn patterns in language and can perform many tasks such as:

- Text generation
- Question answering
- Code generation
- Document summarization
- Translation

Examples of popular LLMs include:

- GPT models
- Claude
- LLaMA
- Gemini

---

### Key Idea

Large Language Models do not simply memorize information.

Instead, they learn **statistical patterns in language**, which allows them to generate coherent and context-aware responses based on the input prompt.

## 3. Overview of the Transformer Architecture

Modern Large Language Models are built using the **Transformer architecture**, which was introduced in the research paper:

> *"Attention Is All You Need" (2017)*

The key innovation of Transformers is the **attention mechanism**, which allows the model to understand relationships between words in a sentence.

---

### Why Transformers Were Needed

Earlier models such as **RNNs and LSTMs** processed text sequentially, meaning they read words one by one.

This caused several problems:

- Slow training
- Difficulty handling long sentences
- Loss of context in long sequences

Transformers solved these problems by allowing the model to **process all words in a sentence simultaneously**.

---

### Tokenization

Before text can be processed by a language model, it must be converted into **tokens**.

Tokens are small units of text such as:

- words
- parts of words
- punctuation

Example:

Sentence:

> "Machine learning is powerful"

Tokenized form: ["Machine", "learning", "is", "powerful"]


Each token is then converted into a **numerical representation called an embedding**.

---

### Attention Mechanism

The most important idea in Transformers is **attention**.

Attention allows the model to determine:

> Which words in the sentence are most relevant to each other.

Example sentence:

> "The animal didn't cross the street because **it** was too tired."

The model must understand that **"it" refers to "animal"**.

Attention allows the model to learn these relationships.

---

### Self-Attention

Transformers use a special type of attention called **self-attention**.

Self-attention allows each word in a sentence to **look at every other word in the sentence** to understand context.

This enables the model to capture:

- relationships between words
- sentence structure
- meaning across long text sequences

---

### Why Transformers Work So Well

Transformers introduced several advantages:

- Parallel processing of text
- Better handling of long context
- Faster training
- Scalability to extremely large models

These advantages made it possible to build **Large Language Models with billions of parameters**.

---

### From Transformers to LLMs

Modern language models such as GPT are built by stacking **many transformer layers together**.

These models are trained on massive text datasets, allowing them to learn patterns in language and generate human-like responses.

## 5. Setting Up the OpenAI API with Python

To build applications using Large Language Models, we need to interact with the model through an API.

In this workshop, we will use the **OpenAI API** to send prompts to a language model and receive generated responses.

The typical workflow is:

User Prompt → API Request → LLM → Generated Response

To do this in Python, we will:

1. Load our API key from environment variables
2. Initialize the OpenAI client
3. Send a prompt to the model
4. Receive and display the generated response

### Load Environment Variables

```python
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Get OpenAI API key
api_key = os.getenv("OPENAI_API_KEY")

print("API key loaded successfully")
```

Please note here two things:
* API keys should never be hardcoded
* .env files keep secrets safe

### Initialize the OpenAI Client

```python
from openai import OpenAI

client = OpenAI()
```

So we have the **client object** is how Python communicates with the OpenAI API.

### First LLM Request

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Explain machine learning in simple terms."}
    ],
)

print(response.choices[0].message.content)
```

You notice here that the model that is being used is "gpt-4o-mini". This tells the API which language model we want to use.

Just like different machine learning algorithms exist, we also have different language models. For e.g.: gpt-40-mini, gpt-40, gpt-4.

We are using gpt-4o-mini because it is fast, low cost and good for experimentation.

### Understanding Message Roles in LLM Conversations

When interacting with a Large Language Model using the API, we send a list called `messages`.
Each message contains a **role** and **content**.

There are three main roles used in LLM conversations:

---

#### 1. System Role

The **system role defines how the AI should behave**.

It acts like instructions that guide the model's personality, tone, or expertise.

Example:
{"role": "system", "content": "You are an expert machine learning teacher."}


This tells the model:

- what type of expert it should behave like
- how it should answer questions

Examples of system instructions:

- You are a Python instructor.
- You are a helpful AI assistant.
- You are a technical interviewer.

---

#### 2. User Role

The **user role represents the human input or prompt**.

This is the question or instruction given to the model.

Example:
{"role": "user", "content": "Explain machine learning in simple terms."}


This tells the model what task it needs to perform.

Examples of user prompts:

- Explain neural networks.
- Generate interview questions.
- Summarize this document.

---

#### 3. Assistant Role

The **assistant role represents the AI's previous response**.

It is mainly used when building **chat-based applications**, where the conversation continues across multiple turns.

Example:
{"role": "assistant", "content": "Machine learning is a field of AI that allows computers to learn patterns from data."}

System sets the rules, User asks the question, Assistant keeps the conversation history.

## Prompt Engineering Basics

Prompt Engineering is the practice of designing effective instructions for Large Language Models to obtain better results.

Unlike traditional programming where we write explicit logic, with LLMs we guide the model using **natural language instructions**.

The quality of the output depends heavily on **how the prompt is written**.

---

### Example 1: Basic Prompt

Prompt:
Explain machine learning


Possible response:

The model generates a general explanation of machine learning.

---

### Example 2: More Specific Prompt

Prompt:
Explain machine learning in simple terms for a beginner.



The response becomes simpler and easier to understand.

---

### Example 3: Structured Prompt

Prompt:
Explain machine learning in 3 bullet points.


Now the output becomes structured.

---

### Key Idea

Better prompts lead to **better outputs**.

A well-designed prompt usually includes:

- Clear instructions
- Context
- Desired output format

## Mini Project: AI Interview Question Generator

Now we will build a simple AI tool that generates interview questions.

This will be the first component of our **AI Interview Preparation Assistant**.

The system will take a topic as input and generate interview questions.

Example:

Input:
Python

Output:
A list of interview questions related to Python.

```python
topic = input("Enter a topic: ")

prompt = f"""
Generate 5 technical interview questions for the topic: {topic}.
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are an expert technical interviewer."},
        {"role": "user", "content": prompt}
    ]
)

print(response.choices[0].message.content)
```
