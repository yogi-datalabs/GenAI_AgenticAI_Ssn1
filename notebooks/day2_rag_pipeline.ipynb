## 1. Limitations of Large Language Models



Large Language Models (LLMs) are extremely powerful, but they also have several important limitations.

Understanding these limitations helps us see **why Retrieval Augmented Generation (RAG) is needed**.

---

### a) Limited Knowledge Cutoff

LLMs are trained on large datasets collected up to a certain point in time.

This means the model **does not know information beyond its training data**.

Example:

If a model was trained on data until 2023, it may not know events that happened in 2024 or later.

---

### b) Hallucination

Sometimes LLMs generate information that **sounds correct but is actually incorrect**.

This is called **hallucination**.

Example:

If you ask an LLM about a research paper that does not exist, it may still generate a convincing answer.

---

### c) Lack of Access to Private Data

LLMs do not automatically have access to:

- company documents
- internal knowledge bases
- private notes
- proprietary datasets

For example, if a company wants an AI assistant to answer questions about internal policies, the LLM cannot access those documents by default.

---

### d) Limited Context Window

LLMs can only process a **limited amount of text at a time**, called the **context window**.

This means large documents cannot be directly fed to the model.

Example:

A 200-page document cannot be processed in a single prompt.

---

### Summary

Because of these limitations, we need a system that can:

- connect LLMs with external knowledge
- retrieve relevant information from documents
- provide accurate and up-to-date answers

This is where **Retrieval Augmented Generation (RAG)** becomes useful.

## 2. What are Embeddings

Embeddings are numerical representations of text that capture the semantic meaning of the content.

Instead of representing text as words, embeddings convert text into vectors (arrays of numbers).

These vectors allow computers to measure how similar two pieces of text are.

For example:

Sentence 1:
"Machine learning is a branch of artificial intelligence."

Sentence 2:
"AI systems can learn patterns from data."

Even though the sentences use different words, they have similar meanings.

When converted into embeddings, their vectors will be close to each other in vector space.

---

### Example Representation

Text:
"Machine learning"

Embedding (vector representation):

[0.12, -0.45, 0.88, 0.33, -0.19, ...]

Each number represents a dimension in a high-dimensional space.

Modern embedding models may produce vectors with hundreds or thousands of dimensions.

For example:

- 384 dimensions
- 768 dimensions
- 1536 dimensions

---

### Why Embeddings Are Useful

Embeddings allow us to perform **semantic search**.

Instead of searching for exact keywords, we can search based on meaning.

Example:

Query:
"How do machines learn from data?"

Relevant document text:
"Machine learning algorithms identify patterns in datasets."

Even though the words are different, embeddings allow the system to recognize that the meanings are similar.

---

### Role of Embeddings in RAG

In a Retrieval Augmented Generation pipeline:

1. Documents are converted into embeddings
2. Embeddings are stored in a vector database
3. A user query is converted into an embedding
4. The system finds the most similar document chunks
5. The LLM generates an answer using those retrieved chunks

## 3. Semantic Search

Traditional search systems rely on **keyword matching**.

This means the search engine looks for exact words that appear in both the query and the document.

Example:

Query:
"machine learning"

Traditional search returns documents that contain the exact words **machine** and **learning**.

However, this approach has limitations.

For example:

Query:
"How do computers learn from data?"

A document containing the text:
"Machine learning algorithms detect patterns in datasets"

might not be retrieved by keyword search because the words are different.

---

### Semantic Search

Semantic search solves this problem by searching based on **meaning rather than exact words**.

Instead of comparing words directly, we compare **embeddings (vector representations)** of text.

Steps in semantic search:

1. Convert documents into embeddings
2. Store these embeddings
3. Convert the user query into an embedding
4. Find the document embeddings that are most similar to the query embedding

The similarity between embeddings is usually measured using **cosine similarity**.

---

### Example

Query:
"How do machines learn from data?"

Document text:
"Machine learning algorithms identify patterns in datasets."

Even though the words are different, semantic search recognizes that the meanings are similar.

---

### Why Semantic Search is Important for RAG

Semantic search allows us to retrieve the **most relevant pieces of information** from large document collections.

These retrieved pieces of text are then provided to the LLM as context to generate accurate answers.

## 4. Vector Databases

A vector database is a specialized database designed to store and search **embeddings (vector representations of data)**.

Since embeddings represent the semantic meaning of text, a vector database allows us to efficiently find **similar pieces of information**.

Unlike traditional databases that store structured data such as numbers and text fields, vector databases store **high-dimensional vectors**.

For example, an embedding might look like this:

[0.12, -0.45, 0.88, 0.33, -0.19, ...]

Each document or text chunk is converted into such a vector and stored in the database.

---

### Why Vector Databases Are Needed

When working with large document collections, we may have **thousands or millions of embeddings**.

To answer a query, we must quickly find the embeddings that are **most similar** to the query embedding.

Vector databases provide efficient algorithms for performing **similarity search**.

Common similarity metrics include:

• Cosine Similarity
• Euclidean Distance
• Dot Product

These metrics help determine how close two embeddings are in vector space.

---

### Popular Vector Databases

Some commonly used vector databases include:

• ChromaDB
• Pinecone
• Weaviate
• FAISS
• Milvus

In this workshop, we will use **ChromaDB**, which is lightweight and easy to integrate with Python applications.

---

### Role of Vector Databases in RAG

In a Retrieval Augmented Generation pipeline:

1. Documents are divided into smaller chunks
2. Each chunk is converted into an embedding
3. Embeddings are stored in a vector database
4. A user query is converted into an embedding
5. The database retrieves the most similar chunks
6. These chunks are passed to the LLM to generate the final answer

## 5. RAG Architecture

Retrieval Augmented Generation (RAG) is a technique that combines **information retrieval** with **language model generation**.

Instead of relying only on the knowledge stored inside the LLM, RAG allows the system to retrieve relevant information from external documents and use it to generate accurate answers.

This approach helps overcome several limitations of LLMs such as hallucination, outdated knowledge, and lack of access to private data.

---

### Basic Idea of RAG

The core idea is simple:

When a user asks a question, the system first **retrieves relevant information from documents**, and then the LLM uses that information to generate an answer.

---

### RAG Pipeline

A typical RAG pipeline works in the following steps:

1. Documents are collected (PDFs, notes, web pages, etc.)
2. Documents are split into smaller chunks
3. Each chunk is converted into an embedding
4. Embeddings are stored in a vector database
5. A user query is converted into an embedding
6. The system retrieves the most relevant document chunks
7. These chunks are sent to the LLM as context
8. The LLM generates the final answer

## 6. Document Ingestion

In a Retrieval Augmented Generation system, the first step is to **load the documents that contain the knowledge we want the AI system to use**.

These documents could be:

• PDF files
• Text documents
• Web pages
• Knowledge base articles
• Research papers

Before we can create embeddings, we first need to **read the documents and extract their text content**.

This process is called **document ingestion**.

---


### Loading the documents now

```python
from langchain_community.document_loaders import PyPDFLoader
import os
```

```python
data_folder = "../data/raw/"

documents = []

for file in os.listdir(data_folder):

    if file.endswith(".pdf"):

        file_path = os.path.join(data_folder, file)

        loader = PyPDFLoader(file_path)

        docs = loader.load()

        documents.extend(docs)

print("Total pages loaded:", len(documents))
```

### Inspect the extracted text

```python
print(documents[1].page_content[:50000])
```

We see here the second page of the first document is read properly.

## 7. Text Chunking

In a Retrieval Augmented Generation system, documents are usually too large to be processed effectively as a whole.

If we directly convert entire documents into embeddings, retrieval may become less accurate because each document can contain many different topics.

To solve this, we split documents into smaller pieces called **chunks**.

---

### Why Chunking is Needed

Chunking helps us:

• improve retrieval quality
• reduce the amount of irrelevant information
• make better use of the LLM context window
• retrieve more precise information

For example, instead of embedding an entire 20-page PDF as one unit, we split it into many smaller chunks.

Then, when the user asks a question, the system retrieves only the most relevant chunks.

---

### Chunking in RAG

The typical process is:

1. Load documents
2. Split them into chunks
3. Convert chunks into embeddings
4. Store the embeddings in a vector database

This makes the retrieval process much more effective.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = text_splitter.split_documents(documents)

print("Total chunks created:", len(chunks))
```

### Inspecting the chunks

```python
print("Chunk 101 - last 200 chars:\n")
print(chunks[101].page_content[-200:])

print("\n" + "="*80 + "\n")

print("Chunk 102 - first 200 chars:\n")
print(chunks[102].page_content[:200])
```

Suppose an important sentence gets split between two chunks. Without overlap, part of the meaning may be lost. With overlap, some shared text appears in both chunks, so retrieval preserves context better.

## 8. Creating Embeddings

After splitting the documents into chunks, the next step is to convert each chunk into an **embedding**.

An embedding is a numerical vector representation of text.

Embeddings capture the semantic meaning of text, which allows us to compare chunks based on meaning rather than exact words.

In a RAG pipeline:

- each chunk is converted into an embedding
- embeddings are stored in a vector database
- the user query is also converted into an embedding
- the system retrieves the most similar chunks using vector similarity

This is the foundation of semantic search.

```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings()

sample_text = chunks[0].page_content

sample_embedding = embedding_model.embed_query(sample_text)

print("Embedding vector length:", len(sample_embedding))
print("First 10 values:", sample_embedding[:10])
```

## 9. Storing Embeddings in ChromaDB

```python
from langchain_community.vectorstores import Chroma

vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=embedding_model,
    persist_directory="../chroma_db"
)
```

```python
print("Number of chunks stored:", vector_db._collection.count())
```

### Checking the Chroma DB

```python
results = vector_db._collection.get(limit=3)

for doc in results["documents"]:
    print(doc)
    print("\n======================================\n")
```

```python
print(results["metadatas"])
```

## 10. Direct LLM Response vs RAG-Based Response

Before using Retrieval Augmented Generation, let us first ask the LLM the same question directly.

This will help us compare:

- a generic answer from the LLM
- a grounded answer generated using retrieved document context

This comparison helps us understand why RAG is useful.

### Checking the llm response

```python
from openai import OpenAI
client = OpenAI()

query = "What is the correct water to rice ratio when cooking Tibetan rice?"

direct_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful cooking assistant."},
        {"role": "user", "content": query}
    ]
)

print("Direct LLM Response:\n")
print(direct_response.choices[0].message.content)
```

### Using RAG now to check the same prompt

```python
# Retrieve documents from Vector DB

retrieved_docs = vector_db.similarity_search(query, k=3)

for i, doc in enumerate(retrieved_docs):
    print(f"\n--- Document {i+1} ---\n")
    print(doc.page_content[:500])

# RAG Prompt

context = "\n\n".join([doc.page_content for doc in retrieved_docs])

rag_prompt = f"""
Answer the question using ONLY the context provided.

Context:
{context}

Question:
{query}
"""
```

### Ask LLM with RAG

```python
rag_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Answer only using the provided context."},
        {"role": "user", "content": rag_prompt}
    ]
)

print(rag_response.choices[0].message.content)
```
